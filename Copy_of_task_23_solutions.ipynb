### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
df = pd.read_csv('/content/List of most-polluted cities by particulate matter concentration - List of most-polluted cities by particulate matter concentration (1).csv')
df.head()
```

```python
iran_series = df.query('Country == "Iran"').groupby('Year')['PM2.5'].sum()
china_series = df.query('Country == "China"').groupby('Year')['PM2.5'].sum()
```

```python
plt.plot(iran_series.index, iran_series.values, label = "Iran", linestyle = 'dashed')
plt.plot(china_series.index, china_series.values, label = "China", linestyle = 'dotted')
plt.title('Chinva VS Iran PM2.5 Yearly comparision')
plt.legend()
plt.xlabel("Year")
plt.ylabel("PM2.5")
plt.show()
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df['PM10'], density=True, bins= 50, facecolor='pink', alpha = 0.6)
plt.xlabel('Bins')
plt.ylabel('Probability')
plt.title('Histogram of PM10')
plt.grid()
plt.show()
```

### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
Chile_series = df.query('Country == "Chile"')
Poland_series = df.query('Country == "Poland"')
```

```python
plt.scatter(Chile_series['PM2.5'],Chile_series.PM10, marker="+",color = 'black', label ="Chile")
plt.scatter(Poland_series['PM2.5'],Poland_series.PM10, marker="o",color = 'yellow', label ="Poland")
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.title('PM2.5 Vs PM10 for Chile and Poland')
plt.legend()
plt.grid()
plt.show()
```

###`Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.


```python
# code here
df['date'] = pd.to_datetime(['2020-{}-01'.format(month) for month in df['month_number']])
```

```python
sales_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer', 'total_units', 'total_profit']
final_df = df.groupby(df['date'].dt.quarter)[sales_columns].sum()
```

```python
final_df
```

```python
plt.plot(final_df.index,final_df.facecream, marker = "o", linestyle = 'dotted', color = 'red', label = 'facecream')
plt.plot(final_df.index,final_df.facewash, marker = "o", linestyle = 'dashdot', color = 'green', label = 'facewash')
plt.plot(final_df.index,final_df.toothpaste, marker = "o", linestyle = 'dotted', color = 'black', label = 'toothpaste')
plt.plot(final_df.index,final_df.bathingsoap, marker = "o", linestyle = 'dashdot', color = 'purple', label = 'bathingsoap')
plt.plot(final_df.index,final_df.shampoo, marker = "o", linestyle = 'dotted', color = 'blue', label = 'shampoo')
plt.plot(final_df.index,final_df.moisturizer, marker = "o", linestyle = 'dashdot', color = 'orange', label = 'moisturizer')


plt.title("Quarterly Product wise sales")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
