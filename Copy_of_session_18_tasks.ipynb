###`Q-6:` Make a dataframe of each team in IPL with details like - Team Name, Matches Played, Win%, Home Win%, Away Win%.
Show sorted dataframe on Win%

Replace old team name as new name before performing any tasks.
```
Delhi Daredevils ->Delhi Capitals
Kings XI Punjab -> Punjab Kings
Rising Pune Supergiants -> Rising Pune Supergiant
```

Note: Team1 represents Home team. Exclude not result matches.


````python
ipl_matches = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv")
ipl_matches.head()
````

````python
ipl_matches['Team1'].replace({"Delhi Daredevils":"Delhi Capitals","Kings XI Punjab":"Punjab Kings","Rising Pune Supergiants":"Rising Pune Supergiant"}, inplace = True)
ipl_matches['Team2'].replace({"Delhi Daredevils":"Delhi Capitals","Kings XI Punjab":"Punjab Kings","Rising Pune Supergiants":"Rising Pune Supergiant"}, inplace = True)
````

````python
ipl_matches['Team1'].unique()
matches = ipl_matches.copy()
````

````python
df = matches[~matches.WinningTeam.isna()]
new_df =pd.DataFrame()
teams = df.Team1.unique()
data = []
for team in teams:
  played = df[(df.Team1 == team) | (df.Team2 == team)].shape[0]
  wins = df[df.WinningTeam == team].shape[0]
  home_wins = df[(df.WinningTeam == team)| (df.Team1 == team)].shape[0] / df[df.Team1 == team].shape[0] * 100
  away_wins = df[(df.WinningTeam == team)| (df.Team2 == team)].shape[0] / df[df.Team2 == team].shape[0] * 100
  data.append([team, played, wins, home_wins, away_wins])
new_df[["Team", "Matches Played", "Wins", "Home Win%", "Away Win%"]] = data
print(new_df.sort_values(by = "Wins", ascending = False, inplace = True))
````

###`Q-8:` Player with most appearance in final match.

`Team1Players` and `Team2Players` have all players name. It is not a list of players name instead it is str. So handle it as string.

Hint: split and strip will help; Make a series of all players in final and do value counts


###`Q-8:` Player with most appearance in final match.

````python
final_matches = ipl_matches[(ipl_matches['MatchNumber'] == 'Final') & (ipl_matches['WonBy'] != 'NoResult')]

all_players = pd.Series(dtype='object')
for index, row in final_matches.iterrows():
    team1_players_str = row['Team1Players']
    team2_players_str = row['Team2Players']

    # Clean and split player names for Team1
    if isinstance(team1_players_str, str):
        team1_players = [p.strip(" '") for p in team1_players_str.strip('[]').split(', ')]
        all_players = pd.concat([all_players, pd.Series(team1_players)])

    # Clean and split player names for Team2
    if isinstance(team2_players_str, str):
        team2_players = [p.strip(" '") for p in team2_players_str.strip('[]').split(', ')]
        all_players = pd.concat([all_players, pd.Series(team2_players)])

most_frequent_player = all_players.value_counts().idxmax()
most_frequent_player_count = all_players.value_counts().max()

print(f"Player with most appearances in final matches: {most_frequent_player}")
print(f"Number of appearances: {most_frequent_player_count}")
````
