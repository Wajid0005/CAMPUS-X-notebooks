### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)

Z = np.abs(X) - np.abs(Y)

fig = plt.figure(figsize= (8,6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface( X,Y,Z, cmap = 'viridis', edgecolor = 'none')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='z value')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('z = |x| - |y|')

plt.tight_layout()
plt.show()
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)

z = np.sin(np.sqrt(xx**2 + yy**2))
````

````python
fig = plt.figure()

ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy,z, cmap = 'magma')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')


plt.show()
````

````python
fig = plt.figure()

ax = plt.subplot()
ax.contour(xx,yy,z, cmap = 'magma')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')


plt.show()
````

````python
fig = plt.figure()

ax = plt.subplot()
ax.contourf(xx,yy,z, cmap = 'magma')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')


plt.show()
````

### `Problem 10:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = tan(\log_2({x^2 + y^2})$$

````python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

X, Y = np.meshgrid(x,y)
z = np.tan(np.log2(X**2 + Y**2))
````

````python
fig = plt.figure()

ax = plt.subplot(projection = '3d')
ax.plot_surface(X,Y,z, cmap = 'plasma')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
````

````python
fig = plt.figure()

ax = plt.subplot()
ax.contour(X,Y,z, cmap = 'magma')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')


plt.show()
````

````python
fig = plt.figure()

ax = plt.subplot()
ax.contourf(X,Y,z, cmap = 'cividis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')


plt.show()
````
